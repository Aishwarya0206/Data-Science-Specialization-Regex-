```python
class checkEqual:
  def __init__(self, num1):
    self.num1 = num1

  def __str__(self):
    return f"Hey my num1 is {self.num1}"

  def __eq__(self, num):
    check = (self.num1 == num.num1)
    return (check)

c1 = checkEqual(10)
c2 = checkEqual(20)
print(c1 == c2)
```

```python
class checkEqual:
  def __init__(self,num1, num2):
    self.num1=num1
    self.num2=num2
 
     
  def __str__(self):
    return f"Hey my num1 is {self.num1}, my num2 is {self.num2}"

  def __eq__(self, temp):
    num1=(self.num1 == temp.num1)
    num2=(self.num2 == temp.num2)
    return checkEqual(num1, num2)

c1 = checkEqual(10, 20)
c2 = checkEqual(20, 20)
print(c1 == c2)
```

```python
def arithmeticResult(value):
  if(value.strip()!=''):
    myfile = open(value+'.txt','w')
    num1 = int(input("Enter number1: "))
    num2 = int(input("Enter number2: "))
    try:
      dic = f"""'Addition': {num1+num2} \n'Subtraction': {num1-num2} \n'Multiplication': {num1*num2}, 
          \n'Float Division': {num1/num2} \n'Modulus': {num1%num2} \n'Integer Division': {num1//num2}"""
      myfile.write(dic)
      myfile.close()
      myfile = open(value+'.txt','r')
      print(myfile.read())
      myfile.close()
    except Exception as ex:
      print(ex)
  else:
    print("File name is not valid")
```

```python
import os
import socket   
from selenium import webdriver

c = 0
while True:
  number = input("Enter a number between 0 & 9: ")
  try:
    if(number.isdigit()):
      if(int(number) in list(range(0,10))):
        if(int(number) == 1):
          location = input("Enter the location: ")
          try:
            if(location.strip!=''):
              os.mkdir(location)
              print("Directory created successfully")
            else:
              print("Enter a proper location")
          except Exception as ex:
            print(ex)
        elif(int(number) == 2):
          try:
            ip_add = socket.gethostbyname(socket.gethostname())
            print("IP Address is",ip_add) 
          except Exception as ex:
            print(ex)
        elif(int(number) == 3):
          try:
            print("Shutdown started...")
            os.system("shutdown /s /t 1")
          except Exception as ex:
            print(ex)
        elif(int(number) == 4):
          try:
            url = "https://docs.python.org/3/library/abc.html"
            wd.get(url)
            print(wd.page_source)
          except Exception as ex:
            print(ex)
        elif(int(number)>=5 and int(number)<=9):
          fileName = input("Enter a file name: ")
          arithmeticResult(fileName)
          print("Successfully written into the file")
        else:
          print("Input can be only within the range 0 to 9")
      else:
        print("Input can be only within the range 0 to 9")
    else:
      c+=1
      if(c<=3):
        print("Input cannot be a string")
        continue
      else:
        print("Error: Provide proper input")
        break
  except Exception as ex:
    print(ex)
```
