```python
df.select_dtypes(object)
```

```python
order = df.groupby(['Vehicle Class']).mean().sort_values('CO2 Emissions(g/km)', ascending = False).index
sns.boxplot(df['CO2 Emissions(g/km)'], df['Vehicle Class'], data = df, order = order)
```

```python
sns.barplot(df['CO2 Emissions(g/km)'], df['Vehicle Class'], data = df, order = order)
```

* VAN - PASSENGER affects CO2 more

```python
order = df.groupby(['Transmission']).mean().sort_values('CO2 Emissions(g/km)', ascending = False).index
plt.figure(dpi=150)
sns.boxplot(df['CO2 Emissions(g/km)'],df['Transmission'], data = df, order = order)
```

```python
plt.figure(dpi=150)
sns.barplot(df['CO2 Emissions(g/km)'],df['Transmission'], data = df, order = order)
```

* Automatic 7 gears transmission emits affects CO2 Emissions

```python
order = df.groupby(['Fuel Type']).mean().sort_values('CO2 Emissions(g/km)', ascending = False).index
plt.figure(dpi=80)
sns.boxplot(df['CO2 Emissions(g/km)'],df['Fuel Type'], data = df, order = order)
```

```python
plt.figure(dpi=80)
sns.barplot(df['CO2 Emissions(g/km)'],df['Fuel Type'], data = df, order = order)
```

* The fuel type E = ethanol (E85) has more effect to the CO2 Emisions

```python
order = df.groupby(['Make']).mean().sort_values('CO2 Emissions(g/km)', ascending = False).index
plt.figure(figsize = (6,10),dpi = 200)
sns.barplot(df['CO2 Emissions(g/km)'],df['Make'], data = df, order = order)
```

* Bugatti cars contribute the most of the CO2 Emissions
